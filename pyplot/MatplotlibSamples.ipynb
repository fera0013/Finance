```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

```

# 1. One dimensional data

```python
np.random.seed(100)
y=np.random.standard_normal(20)
```

```python
x=range(len(y))
```

```python
plt.plot(x,y)
```

```python
plt.plot(y.cumsum())
```

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.axis('tight')
plt.xlim(-1,20)
plt.ylim(np.min(y.cumsum())-1,np.max(y.cumsum())+1)
```

```python
plt.figure(figsize=(7,4))
plt.plot(y.cumsum(),'b',lw=1.5)
plt.plot(y.cumsum(),'ro')
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

# 2. Multidimensional data

```python
np.random.seed(2000)
y=np.random.standard_normal((20,2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(7,4))
plt.plot(y[:,0],lw=1.5,label='1st')
plt.plot(y[:,1],lw=1.5,label='2nd')
plt.plot(y,'ro')
plt.legend(loc=0)
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

## 2.1 Different scales

```python
y[:,0]=y[:,0]*100
plt.figure(figsize=(7,4))
plt.plot(y[:,0],lw=1.5,label='1st')
plt.plot(y[:,1],lw=1.5,label='2nd')
plt.plot(y,'ro')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

## 2.2 Two y-axes (Overlaying subplots)

```python
fig,ax1=plt.subplots()
plt.plot(y[:,0],'b',lw=1.5,label='1st')
plt.legend(loc=0)
plt.plot(y[:,0],'ro')
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')
ax2=ax1.twinx()
plt.plot(y[:,1],'g',lw=1.5,label='2nd')
plt.plot(y[:,1],'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

## 2.3 Two seperate plots

```python
plt.figure(figsize=(7,5))
plt.subplot(211)
plt.plot(y[:,0],'b',lw=1.5,label='1st')
plt.plot(y[:,0],'ro')
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
plt.legend(loc=0)
plt.subplot(212)
plt.plot(y[:,1],'g',lw=1.5,label='2nd')
plt.plot(y[:,1],'ro')
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.legend(loc=0)
```

## 2.4 Different plot types

```python
plt.figure(figsize=(9,4))
plt.subplot(121)
plt.plot(y[:,0],'b',lw=1.5,label='1st')
plt.plot(y[:,0],'ro')
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.legend(loc=0)
plt.title('1st data set')
plt.subplot(122)
plt.bar(np.arange(len(y)),y[:,1],width=0.5,color='g',label='2nd')
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('2nd data set')
```

## 2.5 Scatter plots

```python
y=np.random.standard_normal((1000,2))
```

```python
plt.figure(figsize=(7,5))
plt.plot(y[:,0],y[:,1],'ro')
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
plt.figure(figsize=(7,5))
plt.scatter(y[:,0],y[:,1],marker='o')
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
c=np.random.randint(0,10,len(y))
```

```python
plt.figure(figsize=(7,5))
plt.scatter(y[:,0],y[:,1],c=c,marker='o')
plt.colorbar()
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

## 2.6 Histogram

```python
plt.figure(figsize=(7,5))
plt.hist(y,label=['1st','2nd'],bins=25)
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
```

```python
plt.figure(figsize=(7,5))
plt.hist(y,label=['1st','2nd'],color=['b','g'],stacked=True,bins=20)
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
```

## 2.7 Boxplot

```python
plt.subplots(figsize=(7,4))
plt.boxplot(y)
plt.grid(True)
plt.xlabel('data set')
plt.ylabel('value')
plt.title('Boxplot')
```

# 3. Financial plots

```python
import matplotlib.finance as mpf
```

```python
start=(2014,5,1)
end=(2014,6,30)
```

Get data as list of tuples in the format

[(timestamp_1,open,high,low,close,volume),...(timestamp_n,open,high,low,close,volume)]:

```python
quotes=mpf.quotes_historical_yahoo_ohlc('^GDAXI',start,end)
quotes[:2]
```

## 3.1 Candlestick chart

```python
fig,ax=plt.subplots(figsize=(8,5))
fig.subplots_adjust(bottom=0.2)
mpf.candlestick_ohlc(ax,quotes,width=0.6,colorup='g',colordown='r')
plt.grid(True)
ax.xaxis_date()
ax.autoscale_view()
plt.setp(plt.gca().get_xticklabels(),rotation=30)
```

## 3.2 OHLC-Chart

```python
fig,ax=plt.subplots(figsize=(8,5))
mpf.plot_day_summary_ohlc(ax,quotes,colorup='g',colordown='r')
plt.grid(True)
ax.xaxis_date()
plt.title('DAX index')
plt.ylabel('index level')
plt.setp(plt.gca().get_xticklabels(),rotation=30)
```

## 3.3 Volume

```python
quotes=np.array(mpf.quotes_historical_yahoo_ohlc('YHOO',start,end))
```

```python
fig,(ax1,ax2)=plt.subplots(2,sharex=True,figsize=(8,6))
mpf.candlestick_ohlc(ax1,quotes,width=0.6,colorup='g',colordown='r')
ax1.set_title('Yahoo Inc.')
ax1.set_ylabel('index level')
ax1.grid(True)
ax1.xaxis_date()
plt.bar(quotes[:,0]-0.25,quotes[:,5],width=0.5)
ax2.set_ylabel('volume')
ax2.grid(True)
ax2.autoscale_view()
plt.setp(plt.gca().get_xticklabels(),rotation=30)
```
